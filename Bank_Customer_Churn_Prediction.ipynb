```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

## <font color='yellow'>**Import Dateset**<font>

```python
df=pd.read_csv('Churn_Modelling.csv')
df
```

## <font color='Yellow'>**Data Cleaning**</font>

```python
# Data is Clear
df.info()
```

```python
df["Geography"].unique() # Three countries
```

```python
df.describe()
```

```python
#delete unneeded columns
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
df
```

```python
labelencoder=LabelEncoder()
df['Gender']=labelencoder.fit_transform(df['Gender'])
df['Geography']=labelencoder.fit_transform(df['Geography'])
df
```

```python
# Data Is Ready Now
df.dtypes
```

```python
x=df.drop(columns='Exited') # droping coulmn from the feature
y=df['Exited']
```

## <font color='Blue'>**Logistic Regression Model**</font>

```python
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,shuffle=True,random_state=40)
```

### <font color='orange'>**Train the Model**</font>

```python
LR_model = LogisticRegression()
LR_model.fit(x_train, y_train)
```

### <font color='orange'>**Make Predictions**</font>

```python
y_pred_LR = LR_model.predict(x_test)
```

### <font color='orange'>**Evaluate the Model**</font>

```python
# Accuracy
LR_accuracy = accuracy_score(y_test, y_pred_LR)
print("Accuracy: {:.2f}%".format(LR_accuracy * 100))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred_LR)
print('Confusion Matrix:')
print(conf_matrix)

# Classification Report
class_report = classification_report(y_test, y_pred_LR)
print('Classification Report:')
print(class_report)

# Plot Confusion Matrix
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# visualization
y_prob = LR_model.predict_proba(x_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(10, 5))
plt.plot(fpr, tpr, color='blue', lw=2)
plt.plot([0, 1], [0, 1], color='black', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('LogisticRegression\nAccuracy: {:.2f}%'.format(LR_accuracy * 100))
plt.show()
```

## <font color='Blue'>**Decision Tree Classifier Model**</font>

```python
from sklearn.tree import DecisionTreeClassifier
```

### <font color='orange'>**Train the Model**</font>

```python
clf_model = DecisionTreeClassifier(random_state=42)
clf_model.fit(x_train, y_train)
```

### <font color='orange'>**Make Predictions**</font>

```python
y_pred_clf = clf_model.predict(x_test)
```

### <font color='orange'>**Evaluate the Model**</font>

```python
# Accuracy
clf_accuracy = np.mean(y_pred_clf == y_test)
print("Accuracy: {:.2f}%".format(clf_accuracy * 100))


# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred_clf)
print('Confusion Matrix:')
print(conf_matrix)

# Classification Report
class_report = classification_report(y_test, y_pred_clf)
print('Classification Report:')
print(class_report)

# Plot Confusion Matrix
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# visualization
clf_accuracy = np.mean(y_pred_clf == y_test)
y_prob = clf_model.predict_proba(x_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(10, 5))
plt.plot(fpr, tpr, color='blue', lw=2)
plt.plot([0, 1], [0, 1], color='black', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('DecisionTree\nAccuracy: {:.2f}%'.format(clf_accuracy * 100))
plt.show()
```
